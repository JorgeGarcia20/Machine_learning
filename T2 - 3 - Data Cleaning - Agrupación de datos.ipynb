```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Construcción de un dummy dataset
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500
# Creacion de datos aleatorios para agregarlos al dataset
gender_data = []
income_data = []

for i in range(n):
      gender_data.append(np.random.choice(gender))
      income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
# N( m , s ) => m + s * z
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income": income
    }
)
```

```python
data.head()
```

## Transformación de variables

```python
zscore = lambda x: (x-x.mean())/x.std()
```

```python
z_group = double_group.transform(zscore)
```

```python
plt.hist(z_group["Age"])
```

```python
# Si existieran NaN podrinan ser remplazados con la media de una columna
fill_na_mean = lambda m: m.fillna(m.mean())
```
